```python
#We can populate with any type of chart
g1 = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
g1= g1.map(plt.hist, 'BudgetMillions')
```

```python
#Back to our scatter plot:
g2 = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50, linewidth=0.5, edgecolor ='black')
g2= g2.map(plt.scatter, 'CriticRating' , 'AudienceRating', **kws)
g2.set(xlim=(0,100), ylim=(1,100)) #ths sets the graph boundaries
#graphing a line on each of the facet charts
for ax in g2.axes.flat:
    ax.plot((0,100),(0,100),c='gray', ls="--")
```

```python
#Back to our scatter plot:
g2 = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50, linewidth=0.5, edgecolor ='black')
g2= g2.map(plt.scatter, 'CriticRating' , 'AudienceRating', **kws)
g2.set(xlim=(0,100), ylim=(1,100)) #ths sets the graph boundaries
#graphing a line on each of the facet charts
for ax in g2.axes.flat:
    ax.plot((0,100),(0,100),c='gray', ls="--")
g2.add_legend()
```
